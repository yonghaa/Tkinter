# 배치하기



- pack 함수를 사용하면, 하나의 위젯이 한 줄을 다 차지한다.
- grid 함수를 사용하면, 할당될 수 있는 공간이 미리 나누어져 있다.
- pack과 grid는 기능이 상충하기 때문에 동시에 쓸 수 없다.
- place는 어떠한 공간에 할당하는 것이 아니라, 위젯 자체를 그냥 원하는 위치에 놓는다.

# pack


- side 옵션을 통해 대략적인 위치를 정할 수 있다.(기본=top)
- 위치에 따라 left, right, bottom으로 입력
- padx, pady 옵션을 통해 위젯간의 간격을 줄 수 있다.

- side 옵션을 top으로 지정했다면, 가로줄 전체가 이 위젯에게 할당된 공간이 된다. 그런 경우, padx를 주는 것은 의미가 없다.
    → pady 옵션을 사용함으로서, 아래와 위에 빈 공간을 확보할 수 있다.(위와 아래 공간이 동시에 생김)

- side 옵션을 left 혹은 right로 지정해서, 가로방향으로 배치를 하게 되면, 이 위젯에게 할당되는 공간이 세로줄 전체이다. 그런 경우, pady가 의미 없고 padx를 통해서 위젯의 좌우 간격을 줄 수 있다.

# grid


- 그리드 함수는 엑셀, 표처럼 어떠한 공간이 정해져 있고 그 공간안에 위젯을 배치하는 방식.
## [위치 지정]
- 위치 지정은 가로 방향인 column과 세로 방향인 row를 옵션으로 입력.
- 가로방향인 column은 왼쪽에서 오른쪽, 세로방향 row는 위에서 아래로 방향성을 가진다.
- grid함수를 쓸 때 주의할 점은 **빈공간을 줄 수 없다.**
- ex) 만약 3x3 공간이 지정된 상태에서, 0,4 공간에 위젯을 배치하고 싶다고 적는 경우, 0,3이 아무것도 없기 때문에 강제로 0,3으로 배치된다.
## [공간 병합]
- 공간을 병합할 수도 있다.
- rowspan 옵션 2 → 세로 방향으로 2칸을 병합하겠다.
- columnspan 옵션 2 → 가로 두칸이 병합된다.

# place


## [절대 위치 지정]
- 창 내에서 위젯이 붙을 절대 위치를 지정해 줄 수 있다.
- x, y 옵션으로 위젯이 붙는 위치를 숫자로 지정한다.
- 절대 위치 옵션을 주게 되면 창의 크기가 변하더라도 꿋꿋이 위치를 지키는 특성.
## [상대 위치 지정]
- 창의 크기가 변할 때 위치도 따라서 변하게 하고 싶다면 상대 위치를 지정한다.
- relative(상대적인) 안의 rel을 x,y 앞에 붙인 relx, rely 옵션을 사용한다.
- 창의 길이가 1이라면, 위젯의 위치(왼쪽 위)를 0과 1사이의 값으로 지정한다.
- 창의 크기가 변할때, 위젯의 위치도 그에 따라 변하게 된다.

```python
from tkinter import *  # tkinter에서 모든 함수를 들고 온다.
```

```python
# pack

win = Tk()  # Tk 함수로 창을 만들 수 있다.
win.geometry("400x200")  # geometry 함수를 사용해서 창의 크기 설정.
win.title("pack")  # title함수로 제목 지정
win.option_add("*Font", "궁서 20")  # option_add함수로 글자체 변경 가능.

ent = Entry(win)  # Entry 함수로 입력창 생성
ent.pack()  # pack 함수로 배치, 아무옵션 없이 사용하면 기본값 top에 생성

btn = Button(win)  # Button 함수로 버튼 생성
btn.config(text = "버튼")  # config 함수로 글자를 넣는다.

def aa():
    btn.pack(pady = ent.get())
btn.config(command = aa)  # config함수에 command옵션
btn.pack()  # pack 함수로 배치, 아무옵션 없이 사용하면 기본값 top에 생성

btn2 = Button(win)
btn2.config(text = "temp")
btn2.pack()

win.mainloop()  # 반드시 mainloop 함수를 사용해야 한다.
```

```python
# grid

win = Tk()  # Tk 함수로 창을 만들 수 있다.
win.geometry("400x200")  # geometry 함수를 사용해서 창의 크기 설정.
win.title("grid")  # title함수로 제목 지정
win.option_add("*Font", "궁서 20")  # option_add함수로 글자체 변경 가능.

btn3 = Button(win)  # Button 함수로 만들고, grid 함수를 써서 배치.
btn3.config(text = "(0,0)")
btn3.grid(column = 0, row = 0)  # grid 함수를 쓸 때는, column과 row를 넣어줘야 한다.

btn4 = Button(win)  # Button 함수로 만들고, grid 함수를 써서 배치.
btn4.config(text = "(1,0)")
btn4.grid(column = 1, row = 0)  # grid 함수를 쓸 때는, column과 row를 넣어줘야 한다.

btn5 = Button(win)  # Button 함수로 만들고, grid 함수를 써서 배치.
btn5.config(text = "(0,1)")
btn5.grid(column = 0, row = 1)  # grid 함수를 쓸 때는, column과 row를 넣어줘야 한다.

btn6 = Button(win)  # Button 함수로 만들고, grid 함수를 써서 배치.
btn6.config(text = "(0,3)")
btn6.grid(column = 0, row = 3)  # grid 함수를 쓸 때는, column과 row를 넣어줘야 한다.


win.mainloop()  # 반드시 mainloop 함수를 사용해야 한다.
```

```python
# grid
# 반복문으로 4 x 3 버튼 array 만들자

win = Tk()  # Tk 함수로 창을 만들 수 있다.
win.geometry("600x300")  # geometry 함수를 사용해서 창의 크기 설정.
win.title("grid")  # title함수로 제목 지정
win.option_add("*Font", "궁서 20")  # option_add함수로 글자체 변경 가능.

# 4 x 3 버튼 array 만들자 # 4x3 경우의수 12가지
# 열 두번의 반복 동안 똑같은 변수를 계속 사용하는 것이 문제
# 반복문이 종료되었을 때, btn 변수에는 열 두번째 반복에 사용된 버튼이 저장된다. 나머지는 사라짐.

btn1_list = [] # 리스트를 만들어 놓았고, 사라지기 전에 장바구니(리스트) 안에 담는다.
col_num = 4
row_num = 3  # 4x3 경우의수 12가지
for j in range(0, row_num):  # 세로
    for i in range(0, col_num):  # 가로
        btn1 = Button(win)  
        btn1.config(text = "({},{})".format(i,j))  # format사용. 가로i, 세로j
        btn1.grid(column = i, row = j, padx = 10, pady = 10)  # grid함수에 pad옵션
        btn1_list.append(btn1)

# 병합
btn7 = Button(win)
btn7.config(text = "new")
btn7.grid(column = 1, row = 2, columnspan = 2)  # row방향으로 2칸 엎어서 생성
win.mainloop()  # 반드시 mainloop 함수를 사용해야 한다.
```

```python
# place
# 절대좌표(창의 크기가 변해도 버튼은 그대로있음)
win = Tk()  # Tk 함수로 창을 만들 수 있다.
win.geometry("600x300")  # geometry 함수를 사용해서 창의 크기 설정.
win.title("grid")  # title함수로 제목 지정
win.option_add("*Font", "궁서 20")  # option_add함수로 글자체 변경 가능.

xx = 30
yy = 50

btn8 = Button(win)
btn8.config(text = "({},{})".format(xx,yy))
btn8.place(x = xx, y = yy)


win.mainloop()  # 반드시 mainloop 함수를 사용해야 한다.
```

```python
# place
# 상대좌표(창의 크기가 변하면 버튼도 따라서)
# relx, rely를 써줌 중요한 것은 전체 창의 크기를 1로 가정한다는 점
win = Tk()  # Tk 함수로 창을 만들 수 있다.
win.geometry("600x300")  # geometry 함수를 사용해서 창의 크기 설정.
win.title("place")  # title함수로 제목 지정
win.option_add("*Font", "궁서 20")  # option_add함수로 글자체 변경 가능.

xx = 0.5
yy = 0.5

btn8 = Button(win)
btn8.config(text = "({},{})".format(xx,yy))
btn8.place(relx = xx, rely = yy)


win.mainloop()  # 반드시 mainloop 함수를 사용해야 한다.
```

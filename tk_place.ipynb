```python
from tkinter import *  # tkinter    .
```

```python
# pack

win = Tk()  # Tk     .
win.geometry("400x200")  # geometry     .
win.title("pack")  # title  
win.option_add("*Font", " 20")  # option_add   .

ent = Entry(win)  # Entry   
ent.pack()  # pack  ,     top 

btn = Button(win)  # Button   
btn.config(text = "")  # config   .

def aa():
    btn.pack(pady = ent.get())
btn.config(command = aa)  # config command
btn.pack()  # pack  ,     top 

btn2 = Button(win)
btn2.config(text = "temp")
btn2.pack()

win.mainloop()  #  mainloop   .
```

```python
# grid

win = Tk()  # Tk     .
win.geometry("400x200")  # geometry     .
win.title("grid")  # title  
win.option_add("*Font", " 20")  # option_add   .

btn3 = Button(win)  # Button  , grid   .
btn3.config(text = "(0,0)")
btn3.grid(column = 0, row = 0)  # grid   , column row  .

btn4 = Button(win)  # Button  , grid   .
btn4.config(text = "(1,0)")
btn4.grid(column = 1, row = 0)  # grid   , column row  .

btn5 = Button(win)  # Button  , grid   .
btn5.config(text = "(0,1)")
btn5.grid(column = 0, row = 1)  # grid   , column row  .

btn6 = Button(win)  # Button  , grid   .
btn6.config(text = "(0,3)")
btn6.grid(column = 0, row = 3)  # grid   , column row  .


win.mainloop()  #  mainloop   .
```

```python
# grid
#  4 x 3  array 

win = Tk()  # Tk     .
win.geometry("600x300")  # geometry     .
win.title("grid")  # title  
win.option_add("*Font", " 20")  # option_add   .

# 4 x 3  array  # 4x3  12
#          
#   , btn       .  .

btn1_list = [] #   ,   ()  .
col_num = 4
row_num = 3  # 4x3  12
for j in range(0, row_num):  # 
    for i in range(0, col_num):  # 
        btn1 = Button(win)  
        btn1.config(text = "({},{})".format(i,j))  # format. i, j
        btn1.grid(column = i, row = j, padx = 10, pady = 10)  # grid pad
        btn1_list.append(btn1)

# 
btn7 = Button(win)
btn7.config(text = "new")
btn7.grid(column = 1, row = 2, columnspan = 2)  # row 2  
win.mainloop()  #  mainloop   .
```

```python
# place
# (    )
win = Tk()  # Tk     .
win.geometry("600x300")  # geometry     .
win.title("grid")  # title  
win.option_add("*Font", " 20")  # option_add   .

xx = 30
yy = 50

btn8 = Button(win)
btn8.config(text = "({},{})".format(xx,yy))
btn8.place(x = xx, y = yy)


win.mainloop()  #  mainloop   .
```

```python
# place
# (    )
# relx, rely       1  
win = Tk()  # Tk     .
win.geometry("600x300")  # geometry     .
win.title("place")  # title  
win.option_add("*Font", " 20")  # option_add   .

xx = 0.5
yy = 0.5

btn8 = Button(win)
btn8.config(text = "({},{})".format(xx,yy))
btn8.place(relx = xx, rely = yy)


win.mainloop()  #  mainloop   .
```
